```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips["total_bill"],bins=10)
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")

fig
```

```python
scatter_plot=plt.figure()
axes2=scatter_plot.add_subplot(1,1,1)
axes2.scatter(tips["total_bill"],tips["tip"])
axes2.set_title("Scatter")
axes2.set_xlabel("Total bill")
axes2.set_ylabel("Tip")
```

```python
box_plot=plt.figure()
axes3=box_plot.add_subplot(1,1,1)
axes3.boxplot([tips[tips["sex"]=='Female']['tip'],tips[tips["sex"]=='Male']["tip"]],labels=["Female","Male"])

axes3.set_title("BoxPlot")
axes3.set_xlabel("Sex")
axes3.set_ylabel("Tip")
```

```python
def recode_sex(sex):
    if sex== "Female":
        return 0
    else:
        return 1
```

```python
tips["sex_color"] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot=plt.figure()
axes5=scatter_plot.add_subplot(1,1,1)
```

```python
axes5.scatter(x=tips["total_bill"],y=tips["tip"],s=tips["size"]*10,c=tips['sex_color'],alpha=0.5)
scatter_plot
```
