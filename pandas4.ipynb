```python
import seaborn as sns
```

```python
anscombe=sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
dataset_1=anscombe[anscombe["dataset"]=="I"]
dataset_2=anscombe[anscombe["dataset"]=="II"]
dataset_3=anscombe[anscombe["dataset"]=="III"]
dataset_4=anscombe[anscombe["dataset"]=="IV"]
```

```python
fig = plt.figure()

axes1=fig.add_subplot(2,2,1)
axes2=fig.add_subplot(2,2,2)
axes3=fig.add_subplot(2,2,3)
axes4=fig.add_subplot(2,2,4)

axes1.plot(dataset_1["x"],dataset_1["y"],"o")
axes2.plot(dataset_2["x"],dataset_2["y"],"o")
axes3.plot(dataset_3["x"],dataset_3["y"],"o")
axes4.plot(dataset_4["x"],dataset_4["y"],"o")
```

```python
tips = sns.load_dataset("tips")

print(tips.head())
print(type(tips))
```

```python
fig2=plt.figure()
axes5=fig2.add_subplot(1,1,1)
```

```python
axes5.hist(tips["total_bill"],bins=10)
axes5.set_title("Histogram of Total Bill")
axes5.set_xlabel("Frequency")
axes5.set_ylabel("Total Bill")

fig2
```

```python
scatter_plot=plt.figure()
axes6=scatter_plot.add_subplot(1,1,1)
axes6.scatter(tips["total_bill"],tips["tip"])
axes6.set_title("Scatterplot of Total Bill vs Tip")
axes6.set_xlabel("Total bill")
axes6.set_ylabel("Tip")
```
